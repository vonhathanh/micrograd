```python
# PE(pos, 2i) = sin(pos/10000**(2i/d))
# PE(pos, 2i+1) = cos(pos/10000**(2i/d))
def positional_encoding(pos, d):
    pe = np.zeros(d)
    div_terms = np.exp(np.arange(0, d, 2) * -(np.log(10000.0) / d))
    pe[0::2] = np.sin(pos*div_terms)
    pe[1::2] = np.cos(pos*div_terms)
    return pe
positional_encoding(1, d)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def get_positional_encoding(seq_len, d_model):
    position = np.arange(seq_len)[:, np.newaxis]
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))

    pos_encoding = np.zeros((seq_len, d_model))
    pos_encoding[:, 0::2] = np.sin(position * div_term)
    pos_encoding[:, 1::2] = np.cos(position * div_term)
    
    return pos_encoding

# Generate positional encoding for a sequence
seq_len = 1000  # Length of the sequence
d_model = 1000  # Dimensionality of the model
pos_encoding = get_positional_encoding(seq_len, d_model)
pos_encoding
plt.plot(pos_encoding[999])
plt.show()
# Create a heatmap of positional encodings
# plt.figure(figsize=(10, 8))
# sns.heatmap(pos_encoding, cmap='viridis')
# plt.xlabel('Model Dimensions')
# plt.ylabel('Position (Sequence)')
# plt.title('Positional Encoding Heatmap')
# plt.show()
```
